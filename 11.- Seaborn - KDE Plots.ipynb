```python
#Importar Librerías
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")

%matplotlib inline
```

<h2>KERNEL DENSITY ESTIMATION PLOT - PROBABILITY DENSITY FUNCTION (PDF)</h2>

<h3>Crear un dataset para modelo KDE </h3>

```python
mean = [0,0]
cov = [[0.2,0], [0,3]]
```

```python
x_axis, y_axis = np.random.multivariate_normal(mean, cov, size=40).T
```

## Gráfica de Valores X

```python
sns.kdeplot(x_axis)
```

## Gráfica de Valores Y

```python
sns.kdeplot(y_axis)
```

### Agregar sombra de relleno

```python
sns.kdeplot(x_axis, shade=True)
```

### Cambiar color

```python
sns.kdeplot(x_axis, shade=True, color = "red")
```

## Gráfica KDE Plot

```python
sns.kdeplot(x_axis, y_axis, color = "purple")
```

### Agregar sombra

```python
sns.kdeplot(x_axis, y_axis, shade=True, color = "red")
```

### Agregar paleta de colores

```python
sns.kdeplot(x_axis, y_axis, shade=False, cmap="magma")
```

### Aumentar las líneas

```python
sns.kdeplot(x_axis, y_axis, shade=False, cmap="magma", n_levels=25)
```

### Bandwith 

```python
sns.kdeplot(y_axis, bw=0.5)
```

### Cambiar orientación de gráfica

```python
sns.kdeplot(x_axis, vertical = True)
```

```python
#Cargar dataset
iris= sns.load_dataset("iris")
```

```python
#Asignar variables
setosa = iris.loc[iris.species == "setosa"]
versicolor = iris.loc[iris.species == "versicolor"]
```

## Graficar múltiples KDE Plots

```python
sns.kdeplot(setosa.petal_length, setosa.petal_width)
sns.kdeplot(versicolor.petal_length, versicolor.petal_width)
```

### Agregar paleta de colores

```python
sns.kdeplot(setosa.petal_length, setosa.petal_width, cmap="winter_r" , )
sns.kdeplot(versicolor.petal_length, versicolor.petal_width, cmap="jet")
```

### Agregar paleta de colores, sombra y líneas

```python
sns.kdeplot(setosa.petal_length, setosa.petal_width, cmap="winter_r" , )
sns.kdeplot(versicolor.petal_length, versicolor.petal_width, cmap="jet", shade=True, n_levels=35)
```

### Retirar sombra

```python
sns.kdeplot(setosa.petal_width, setosa.petal_length, cmap="winter_r" , )
sns.kdeplot(versicolor.petal_width, versicolor.petal_length, cmap="jet", shade=False, n_levels=35)
```
