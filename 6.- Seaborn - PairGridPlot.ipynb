```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")

%matplotlib inline
```

```python
#Carga de dataset
iris = sns.load_dataset("iris")
```

```python
#Previsualizar dataset
iris.head()
```

## Agregar Pair Grid - Scatter referente a Dataset

```python
x = sns.PairGrid(iris)
x.map(plt.scatter)
```

## Agregar Pair Grid - Histograma Diagonal

```python
sns.set(style="darkgrid")
x = sns.PairGrid(iris)
x.map_diag(plt.hist)
```

## Agregar Pair Grid - Histograma Diagonal y Scatter Plots para los otros gr√°ficos

```python
sns.set(style="darkgrid")
x = sns.PairGrid(iris)
x.map_diag(plt.hist)
x.map_offdiag(plt.scatter)
```

## Agregar segmentaci√≥n a los datos

```python
sns.set(style="darkgrid")
x = sns.PairGrid(iris, hue="species")
x.map_diag(plt.hist)
x.map_offdiag(plt.scatter)
```

<h3>Agregar legenda a gr√°fico </h3>

```python
sns.set(style="darkgrid")
x = sns.PairGrid(iris, hue="species")
x.map_diag(plt.hist)
x.map_offdiag(plt.scatter)
x = x.add_legend()
```

## Cambiar paleta de colores

```python
sns.set(style="darkgrid")
x = sns.PairGrid(iris, hue="species", palette = "coolwarm")
x.map_diag(plt.hist)
x.map_offdiag(plt.scatter)
x = x.add_legend()
```

## Cambiar estilo de Histograma

```python
sns.set(style="darkgrid")
x = sns.PairGrid(iris, hue="species", palette = "coolwarm")
x.map_diag(plt.hist, histtype="step")
x.map_offdiag(plt.scatter)
x = x.add_legend()
```

## Cambiar estilo de linea de histograma

```python
sns.set(style="darkgrid")
x = sns.PairGrid(iris, hue="species", palette = "coolwarm")
x.map_diag(plt.hist, histtype="step", linewidth=2)
x.map_offdiag(plt.scatter)
x = x.add_legend()
```

## PairGrid de Dos Variables

```python
x = sns.PairGrid(iris, vars=["petal_length","petal_width"])
x = x.map(plt.scatter)
```

## Cambiar diagonal de scatter a histograma

```python
x = sns.PairGrid(iris, vars=["petal_length","petal_width"])
x = x.map_diag(plt.hist)
x = x.map_offdiag(plt.scatter)
```

## Agregar segmentaci√≥n de datos y leyenda al gr√°fico

```python
x = sns.PairGrid(iris, hue="species", vars=["petal_length","petal_width"])
x = x.map_diag(plt.hist)
x = x.map_offdiag(plt.scatter)
x = x.add_legend()
```

## Cambiar color de l√≠nea de contorno

```python
x = sns.PairGrid(iris, hue="species", vars=["petal_length","petal_width"])
x = x.map_diag(plt.hist, edgecolor="red")
x = x.map_offdiag(plt.scatter, edgecolor = "black")
x = x.add_legend()
```

## Pair Grid Multiples Variables

```python
x = sns.PairGrid(iris, hue="species", x_vars=["petal_length","petal_width"], y_vars=["sepal_length","sepal_width"])
x = x.map(plt.scatter, edgecolor = "black")
```

## PairPlot Dataset

```python
x = sns.PairGrid(iris)
x = x.map_diag(plt.hist, edgecolor="black")
x = x.map_upper(plt.scatter)
x = x.map_lower(sns.kdeplot)
```

## Agregar segmentaci√≥n y leyenda

```python
x = sns.PairGrid(iris, hue = "species")
x = x.map_diag(plt.hist, edgecolor="black")
x = x.map_upper(plt.scatter)
x = x.map_lower(sns.kdeplot)
x = x.add_legend()
```

## Cambiar los marcadores de las series de datos

```python
x = sns.PairGrid(iris,hue="species", hue_kws={"marker":["D","s","+"]})
x = x.map(plt.scatter)
x = x.add_legend()
```

## Cambiar tama√±o de marcadores

```python
x = sns.PairGrid(iris,hue="species", hue_kws={"marker":["D","s","+"]})
x = x.map(plt.scatter, s=30)
x = x.add_legend()
```
