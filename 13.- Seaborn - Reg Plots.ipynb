```python
#Importar librerías
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")
#from scipy.stats import spearmanr

%matplotlib inline
```

```python
#Cargar dataset
tips= sns.load_dataset("tips")
```

```python
#Previsualizar dataset 
tips.head()
```

## Reg Plots

```python
sns.regplot(x="total_bill", y="tip", data=tips, color="orange")
```

```python
mean=[2,5]
cov=[[1.1,0.4], [2.2,3]]
x_value, y_value = np.random.multivariate_normal(mean, cov, 100).T
```

```python
sns.regplot(x=x_value, y=y_value, color="red")
```

### Crear con Pandas , series de datos para asignar nombre a los ejes

```python
seriesx_values = pd.Series(x_value, name="Var_X")
seriesy_values = pd.Series(y_value, name="Var_Y")
```

## REG Plot

```python
sns.regplot(x=seriesx_values, y=seriesy_values)
```

### Cambiar tipo de marcador

```python
sns.regplot(x=seriesx_values, y=seriesy_values, marker="d")
```

```python
sns.regplot(x="total_bill", y="tip", data=tips, color="blue")
```

### Ajustar color y ancho de línea de regresión

```python
sns.regplot(x="total_bill", y="tip", data=tips, scatter_kws={"marker":"d", "color":"blue"} , line_kws={"color":"yellow", "linewidth":5})
```

### Ajustar color y ancho de línea de regresión y marcadores

```python
sns.regplot(x="total_bill", y="tip", data=tips, scatter_kws={"marker":"o", "color":"gray"} , line_kws={"color":"black", "linewidth":2})
```

```python
sns.regplot(x=x_value, y=y_value, data=tips, scatter_kws={"marker":"o", "color":"purple"} , line_kws={"color":"green", "linewidth":2})
```

### Ajustar el intervalo de confianza de línea de regresión

```python
sns.regplot(x=x_value, y=y_value, data=tips , ci=100
           )
```

### Ajustar separación de puntos

```python
sns.regplot(x="size", y="total_bill", data=tips , ci=100, x_jitter=0.2, color = "black"
           )
```

```python
sns.regplot(x="size", y="total_bill", data=tips , ci=100, x_jitter=0.2, color = "black", scatter_kws={"color":"yellow"}, line_kws = {"color":"green"}
           )
```

### Agregar estimador de Intervalo de Confianza

```python
sns.regplot(x="size", y="total_bill", data=tips, x_estimator=np.mean, line_kws={"color":"black"}, scatter_kws={"color":"red"})
```
