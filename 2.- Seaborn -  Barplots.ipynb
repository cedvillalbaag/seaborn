# DataVisualization - BarPlots

```python
#Importar dataset
tips = sns.load_dataset("tips")
sns.barplot(x="day",y="tip", data=tips)
```

### Agregar segmentación de data por sexo

```python
tips = sns.load_dataset("tips")
sns.barplot(x="day",y="tip", data=tips, hue="sex")
```

### Cambiar paleta de colores

```python
tips = sns.load_dataset("tips")
sns.barplot(x="day",y="tip", data=tips, hue="sex", palette="winter_r")
```

### Cambiar variables de análisis

```python
sns.barplot(x="day",y="total_bill", data=tips, hue="smoker")
```

### Cambiar Gráfico a Barras Horizontales (Cambiar datos en las variables X por Y) 

```python
sns.barplot(x="total_bill",y="day", data=tips, hue="smoker")
```

### Cambiar nuevamente paleta de colores

```python
sns.barplot(x="total_bill",y="day", data=tips, hue="smoker", palette= "spring")
```

### Cambiar la posición de las barras - Eje X

```python
sns.barplot(x="day",y="tip", data=tips, palette = "spring", order = ["Sat","Sun","Thur","Fri"])
```

### Grafico Original nuevamente

```python
sns.barplot(x="day", y="total_bill", data=tips)
```

### Agregar un estimador (Mean, Median) a la data del grafico

```python
from numpy import median
sns.barplot(x="day",y="total_bill", data=tips, estimator=median, palette = "spring")
```

```python
sns.barplot(x="smoker",y="tip", data=tips, estimator=median, palette="coolwarm")
```

# Segunda Parte

```python
sns.set(style="darkgrid")
sns.barplot(x="smoker",y="total_bill", data=tips, estimator=median, hue="sex", palette="coolwarm")
```

### Cambiar intervalo de confianza

```python
sns.set(style="darkgrid")
sns.barplot(x="smoker",y="total_bill", data=tips, ci=58, hue="sex", palette="coolwarm")
```

### Cambiar forma del estimador intervalo de confianza

```python
sns.set(style="darkgrid")
sns.barplot(x="smoker",y="total_bill", data=tips, ci=68, capsize=0.1, hue="sex", palette="autumn",  )
```

### Definir Color de Barras Uniforme

```python
sns.set(style="darkgrid")
sns.barplot(x="size",y="tip", data=tips, ci=68, color="blue", capsize=0.03 )
```

### Cambiar Saturación del Color Barras

```python
sns.set(style="darkgrid")
sns.barplot(x="size",y="tip", data=tips, ci=68, color="blue", capsize=0.025, saturation = 0.2 )
```
