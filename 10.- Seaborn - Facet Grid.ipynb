```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")

%matplotlib inline
```

```python
#Cargar dataset
tips = sns.load_dataset("tips")
```

```python
#Previsualizar dataset
tips.head()
```

```python
#Previsualizar dataset
tips.tail()
```

```python
#Análisis estadístico
tips.describe()
```

## Facet Grid

```python
sns.FacetGrid(tips, row="smoker", col="time")
```

## Facet Grid - Histogramas

```python
x = sns.FacetGrid(tips, row="smoker", col="time")
x = x.map(plt.hist, "total_bill")
```

### Cambiar color histograma

```python
x = sns.FacetGrid(tips, row="smoker", col="time")
x = x.map(plt.hist, "total_bill", color = "green")
```

### Agregar más barras al gráfico

```python
x = sns.FacetGrid(tips, row="smoker", col="time")
x = x.map(plt.hist, "total_bill", color = "green", bins=30)
```

## Facet grid - Scatter Plot

```python
x = sns.FacetGrid(tips, row="smoker", col="time")
x = x.map(plt.scatter, "total_bill", "tip")
```

## Facet Grid - Scatter Plot con Regresión

```python
x = sns.FacetGrid(tips, row="smoker", col="time")
x = x.map(sns.regplot, "total_bill", "tip")
```

## Facet Grid - Scatter - Una Columna

```python
x = sns.FacetGrid(tips, row="smoker")
x = x.map(plt.scatter, "total_bill", "tip")
```

## Agregar segmentación de data | Cambiar orientación

```python
x = sns.FacetGrid(tips, col="time", hue="smoker")
x = x.map(plt.scatter, "total_bill", "tip")
x = x.add_legend()
```

## Retirar leyenda

```python
x = sns.FacetGrid(tips, col="day")
x = x.map(plt.scatter, "total_bill", "tip")
x = x.add_legend()
```

## Agregar leyenda con segmentación

```python
x = sns.FacetGrid(tips, col="day", hue="smoker")
x = x.map(plt.scatter, "total_bill", "tip")
x = x.add_legend()
```

## Facet Grid - Boxplot

```python
x = sns.FacetGrid(tips, col="day")
x = x.map(sns.boxplot,"time" ,"total_bill" )
x = x.add_legend()
```

## Ajustar tamaño de gráfico

```python
x = sns.FacetGrid(tips, col="day", size=2)
x = x.map(sns.boxplot,"time" ,"total_bill" )
x = x.add_legend()
```

## Ajustar tamaño de gráfico

```python
x = sns.FacetGrid(tips, col="day", size=7, aspect= 0.4)
x = x.map(sns.boxplot,"time" ,"total_bill" )
x = x.add_legend()
```

## Agregar orden de columna

```python
x = sns.FacetGrid(tips, col="day", col_order = ["Sat", "Sun", "Thur","Fri"], size=7, aspect= 0.4)
x = x.map(sns.boxplot,"time" ,"total_bill" )
x = x.add_legend()
```

## Cambiar color

```python
x = sns.FacetGrid(tips, col="day", col_order = ["Sat", "Sun", "Thur","Fri"], size=7, aspect= 0.4)
x = x.map(sns.boxplot,"time" ,"total_bill", color="red")
x = x.add_legend()
```

```python
x = sns.FacetGrid(tips, col="day", col_order = ["Sat", "Sun", "Thur","Fri"], size=7, aspect= 0.4)
x = x.map(sns.boxplot,"time" ,"total_bill" )
x = x.add_legend()
```

## Facet Grid - Scatter Plot con segmentación de datos y paleta de colores

```python
x = sns.FacetGrid(tips, row="smoker", hue="time", palette="winter_r")
x = x.map(plt.scatter, "total_bill", "tip")
x = x.add_legend()
```

### Ajustar tamaño de gráfico

```python
x = sns.FacetGrid(tips, row="smoker", hue="time", palette="winter_r", size=4)
x = x.map(plt.scatter, "total_bill", "tip")
x = x.add_legend()

```

### Ajustar tamaño de gráfico

```python
x = sns.FacetGrid(tips, row="smoker", hue="time", palette="winter_r", size=2, aspect=4)
x = x.map(plt.scatter, "total_bill", "tip")
x = x.add_legend()
```
