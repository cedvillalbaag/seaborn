```python
#Importar librerías
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")
#from scipy.stats import spearmanr

%matplotlib inline
```

```python
#Cargar dataset 
tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
```

```python
# Previsualizar dataset
tips.head()
```

```python
# Previsualizar dataset 
iris.head()
```

## Gráfica PairPlot

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris, palette="magma", hue="species")
```

```python
sns.pairplot(tips, palette="summer", hue="smoker")
```

```python
sns.pairplot(tips, palette="Blues_d", hue="smoker")
```

```python
sns.pairplot(iris, palette="magma", hue="species", markers=["o","d","s"]) #+
```

```python
sns.pairplot(iris, palette="magma", hue="species", vars=["sepal_length","sepal_width"])
```

```python
sns.pairplot(tips, palette="magma", hue="day", vars=["total_bill","tip"])
```

```python
sns.pairplot(iris, palette="magma", hue="species",  vars=["sepal_length","sepal_width"], size=5)
```

```python
sns.pairplot(iris, x_vars=["petal_length", "petal_width"], palette="magma", hue="species",  y_vars=["sepal_length","sepal_width"], size=3, markers=["o","d","s"])
```

```python
sns.pairplot(iris, palette="magma", hue="species", diag_kind="kde")
```

```python
sns.pairplot(iris, palette="magma", hue="species", kind="reg")
```

```python
sns.pairplot(iris, palette="magma", hue="species", kind="reg", diag_kind="kde")
```

```python
sns.pairplot(tips, palette="magma", hue="smoker", kind="reg")
```
