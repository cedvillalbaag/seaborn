```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#sns.set(style="darkgrid")

%matplotlib inline
```

```python
# Cargar el dataset
tips = sns.load_dataset("tips")
```

```python
# Previsualizar dataset
tips.head(10)
```

```python
# Información Estadística de Dataset
tips.describe()
```

## Strip Plot - Para una columna

```python
sns.stripplot(x=tips["tip"])
```

## Cambio de Color

```python
sns.set(style="darkgrid")
sns.stripplot(x=tips["tip"] , color = "red")
```

## Cambio de Variable

```python
sns.stripplot(x=tips["size"])
```

```python
sns.set(style="darkgrid")
sns.stripplot(x=tips["total_bill"] , color = "red")
```

## Strip Plot para dos variables (X y Y) del dataset | Uso de Paleta de Colores

```python
sns.set(style="darkgrid")
sns.stripplot(x="day",y="total_bill", data=tips, palette = "husl")
```

## Quitar separación de Puntos (Jitter)

```python
sns.set(style="darkgrid")
sns.stripplot(x="day",y="total_bill", data=tips, palette = "husl", jitter=False)
```

## Ajustando Separación de Puntos (Jitter)

```python
sns.set(style="darkgrid")
sns.stripplot(x="day",y="total_bill", data=tips, palette = "husl", jitter=0.1)
```

## Cambiar a orientación Horizontal (Intercambiar variables X y Y)

```python
sns.set(style="darkgrid")
sns.stripplot(x="total_bill",y="day", data=tips, palette = "husl", jitter=0.1)
```

## Agregar línea de contorno a los marcadores

```python
sns.set(style="darkgrid")
sns.stripplot(x="total_bill",y="day", data=tips, palette = "husl", jitter=0.1, linewidth = 1.9)
```

## Agregar segmentación de datos

```python
sns.set(style="darkgrid")
sns.stripplot(x="day",y="total_bill", data=tips, hue="sex", palette = "husl", jitter=0.1, linewidth = 1.9)
```

## Agregar separación a la segmentación de datos (Hue)

```python
sns.set(style="darkgrid")
sns.stripplot(x="day",y="total_bill", data=tips, hue="sex", palette = "husl", jitter=0.1, linewidth = 1.9, dodge=True)
```

## Cambiar variable de segmentación de datos (Hue)

```python
sns.set(style="darkgrid")
sns.stripplot(x="day",y="total_bill", data=tips, hue="smoker", palette = "winter_r", jitter=0.1, linewidth = 1.9, dodge=True)
```

## Cambiar el tipo de marcador

```python
sns.set(style="darkgrid")
sns.stripplot(x="day",y="total_bill", data=tips, hue="smoker", palette = "winter_r", jitter=0.1, linewidth = 1.1, dodge=True, marker = "d")
```

## Cambiar el tamaño del marcador

```python
sns.set(style="darkgrid")
sns.stripplot(x="day",y="total_bill", data=tips, hue="smoker", palette = "winter_r", jitter=0.1, linewidth = 1.1,dodge=True, marker = "d", size=10)
```

## Cambiar color a la línea de contorno de los marcadores

```python
sns.set(style="darkgrid")
sns.stripplot(x="day",y="total_bill", data=tips, hue="smoker", palette = "winter_r", jitter=0.1, linewidth = 1.1,dodge=True, marker = "d", size=10, edgecolor="gray")
```

## Agregar opacidad a la línea de contorno

```python
sns.set(style="darkgrid")
sns.stripplot(x="day",y="total_bill", data=tips, hue="smoker", palette = "winter_r", jitter=0.1, linewidth = 1.1,dodge=True, marker = "d", size=10, edgecolor="red", alpha=0.4)
```

## Combinar Stripplot con Boxplot

```python
sns.stripplot(x="tip", y="day", data =tips,palette="winter_r", color=1)
sns.boxplot(x="tip", y="day", data=tips)
```

## Combinar stripplot con violinplot

```python
sns.stripplot(x="tip", y="day", data =tips,palette="winter_r", color=1)
sns.violinplot(x="tip", y="day", data=tips, palette = "spring")
```

## Ajustar color de violinplot

```python
sns.stripplot(x="tip", y="day", data =tips,palette="winter_r", color=1)
sns.violinplot(x="tip", y="day", data=tips, color="0.4")
```

## Quitar color relleno violinplot

```python
sns.stripplot(x="tip", y="day", data =tips,palette="winter_r", color=1)
sns.violinplot(x="tip", y="day", data=tips, color="0.9")
```

## Ajustar color relleno violinplot

```python
sns.stripplot(x="tip", y="day", data =tips,palette="winter_r", color=1)
sns.violinplot(x="tip", y="day", data=tips, color="0.2")
```
