```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")

%matplotlib inline
```

```python
# Cargar Dataset
tips = sns.load_dataset("tips")
```

```python
# Previsualizar Data
tips.head(10)
```

```python
# Análisis estadísticos de la data
tips.describe()
```

## Violin Plot  (Una Columna)


```python
sns.violinplot(x=tips["tip"])
```

```python
sns.violinplot(x=tips["total_bill"])
```

## Violin Plot para Dos Variables

```python
sns.violinplot(x="day", y="total_bill",data=tips)
```

```python
sns.violinplot(x="size", y="total_bill",data=tips)
```

### Agregar paleta de colores

```python
sns.violinplot(x="day", y="total_bill",data=tips, palette="winter")
```

## Cambiar paleta de colores y agregar segmentación

```python
sns.violinplot(x="day", y="total_bill",data=tips, palette="summer", hue="smoker")
```

```python
sns.violinplot(x="day", y="total_bill",data=tips, palette="spring", hue="sex")
```

```python
sns.violinplot(x="day", y="total_bill",data=tips, palette="spring", hue="smoker")
```

### Unir en el mismo elemento la segmentación de la data

```python
sns.violinplot(x="day", y="total_bill",data=tips, palette="spring", hue="smoker", split=True)
```

### Cambiar el orden de las columnas

```python
sns.violinplot(x="day", y="total_bill",data=tips, palette="spring", hue="smoker", split=True, order =["Fri","Sat","Sun","Thur"])
```

## Agregar Cuartiles a los elementos

```python
sns.violinplot(x="day", y="total_bill",data=tips, hue="smoker", palette="coolwarm",inner="quartile")
```

### Unir segmentación en un mismo elemento

```python
sns.violinplot(x="day", y="total_bill",data=tips, hue="sex", palette="coolwarm",inner="quartile", split=True)
```

```python
sns.violinplot(x="day", y="total_bill",data=tips, hue="sex", palette="coolwarm",inner="quartile", split=True, scale="count")
```

```python
sns.violinplot(x="day", y="total_bill",data=tips, hue="smoker", palette="coolwarm",inner="quartile", scale="count")
```

### Cambiar alineación a horizontal

```python
sns.violinplot(x="total_bill", y="day",data=tips, hue="smoker", palette="coolwarm",inner="quartile", scale="count")
```

```python
sns.violinplot(x="day", y="total_bill",data=tips, hue="smoker", palette="coolwarm",inner="stick", scale="count")
```

```python
sns.violinplot(x="day", y="total_bill",data=tips, hue="smoker", palette="coolwarm",inner="stick")
```

```python
sns.violinplot(x="day", y="total_bill",data=tips, hue="sex", palette="husl",inner="stick", scale="count" , split = True)
```

```python
sns.violinplot(x="day", y="total_bill",data=tips, hue="sex", palette="husl",inner="stick", scale="count", split = True, scale_hue=False)
```

```python
sns.violinplot(x="day", y="total_bill",data=tips, hue="sex", palette="husl",inner="stick", scale="count", split = True, scale_hue=False, bw=0.2)
```
