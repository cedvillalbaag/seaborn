```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")
from scipy.stats import spearmanr

%matplotlib inline
```

<h3> Cargar Dataset </h3>

```python
tips = sns.load_dataset("tips")
```

```python
iris = sns.load_dataset("iris")
```

```python
#Previsualizar dataset
tips.head()
```

```python
#Previsualizar dataset
iris.head()
```

## Joint Plot

```python
sns.jointplot(x="total_bill", y="tip", data=tips)
```

```python
x = sns.jointplot(x="sepal_length", y="sepal_width", data=iris)
```

### Agregar regresión

```python
sns.jointplot(x="total_bill", y="tip", data=tips, kind="reg")
```

### Cambiar color

```python
sns.jointplot(x="total_bill", y="tip", data=tips, kind="reg", color="red")
```

### Cambiar marcador de scatter plot

```python
sns.jointplot(x="total_bill", y="tip", data=tips, kind="hex", color="red")
```

```python
sns.jointplot(x="total_bill", y="tip", data=tips, kind="kde", color="red")
```

```python
x = sns.jointplot(x="sepal_length", y="sepal_width", data=iris, kind = "reg")
```

```python
x = sns.jointplot(x="sepal_length", y="sepal_width", data=iris, kind="kde")
```

### Agregar coeficiente de spearmanr

```python
sns.jointplot(x="total_bill", y="tip", data=tips, kind="reg", color="red", stat_func=spearmanr)
```

```python
sns.jointplot(x="size", y="total_bill", data=tips, kind="reg", color="red", stat_func=spearmanr)
```

### Ajustar tamaño

```python
sns.jointplot(x="total_bill", y="tip", data=tips, kind="reg", color="red", ratio=10, size=6)
```
