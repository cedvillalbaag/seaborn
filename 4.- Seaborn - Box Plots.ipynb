```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
#Cargar Dataset
tips = sns.load_dataset("tips")
```

```python
#Previsualizar Dataset
tips.head(10)
```

```python
#Previsualizar Dataset
tips.describe()
```

## Graficar Boxplot para una columna de la data

```python
sns.boxplot(x = tips["size"])
```

```python
sns.boxplot(x = tips["total_bill"])
```

## Graficar Boxplot para Dos Variables

```python
sns.boxplot(x="sex", y="total_bill", data=tips)
```

```python
sns.boxplot(x="day", y="total_bill", data=tips)
```

## Agregar segmentación de datos y paleta de colores

```python
sns.boxplot(x="day", y="total_bill", data=tips, hue = "sex", palette = "husl")
```

```python
sns.boxplot(x="day", y="total_bill", data=tips, hue = "smoker", palette = "coolwarm")
```

```python
sns.boxplot(x="day", y="total_bill", data=tips, hue = "time")
```

```python
#Cargar Dataset
iris = sns.load_dataset("iris")
```

```python
#Previsualizar Dataset
iris.head(10)
```

## Box plot aplicado a Dataset

```python
sns.boxplot(data=iris)
```

## Cambiar dirección de cajas a horizontal

```python
sns.boxplot(data=iris, orient="h")
```

## Cambiar dirección de cajas a vertical

```python
sns.boxplot(data=iris, orient="v")
```

## Agregar otro tipo de Gráfico: DataPoint - Swarm Plot

```python
sns.boxplot(x="day", y="total_bill", data=tips)
sns.swarmplot(x="day", y="total_bill", data=tips)
```

## Cambiar atributos de Swarm Plot - Color

```python
sns.boxplot(x="day", y="total_bill", data=tips, palette="spring")
sns.swarmplot(x="day", y="total_bill", data=tips, color="black")
```

## Cambiar atributos de Swarm Plot - Color

```python
sns.boxplot(x="day", y="total_bill", data=tips, palette="spring")
sns.swarmplot(x="day", y="total_bill", data=tips, color="0.6")
```
