```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")

%matplotlib inline
```

```python
#Cargar Dataset
flights = sns.load_dataset("flights")
```

```python
#Previsualizar dataset
flights.head()
```

```python
#Análisis estadístico de la data
flights.describe()
```

```python
# Pivotear columnas
flights = flights.pivot("month","year","passengers")
```

```python
# Previsualizar Data
flights.head()
```

## Heat Map - Dataset(Dos Variables)

```python
sns.heatmap(flights)
```

## Cluster Map - Dataset (Dos Variables)

```python
sns.clustermap(flights)
```

### Quitar cluster de columnas

```python
sns.clustermap(flights, col_cluster =False)
```

### Quitar cluster de filas

```python
sns.clustermap(flights, row_cluster =False)
```

### Agregar paleta de colores

```python
sns.clustermap(flights, cmap="coolwarm")
```

### Espacio entre filas y columnas

```python
sns.clustermap(flights, cmap="coolwarm", linewidths=3)
```

### Cambio de paleta de colores

```python
sns.clustermap(flights, cmap="Blues_r", linewidths=3)
```

### Cambio de paleta de colores y tamaño de figura

```python
sns.clustermap(flights, cmap="coolwarm", linewidths=3, figsize=(16,4))
```
