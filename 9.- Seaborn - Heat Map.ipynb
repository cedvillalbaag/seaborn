```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")

%matplotlib inline
```

## Crear un arreglo de datos

```python
normal= np.random.rand(12,15)
```

```python
normals = pd.DataFrame(normal)
```

```python
normals.head()
```

## Heat Map - Grafica de un Arreglo (2 variables)

```python
sns.heatmap(normal)
```

```python
#Serie Pandas
sns.heatmap(normals)
```

## Agregar etiquetas de datos

```python
sns.heatmap(normal, annot= True)
```

## Ajustar escala de colores

```python
sns.heatmap(normal, annot= True, vmin=0, vmax=2)
```

```python
#Cargar dataset flights
flights = sns.load_dataset("flights")
```

```python
#Previsualizar dataset
flights.head()
```

```python
#Previsualizar dataset
flights.tail()
```

```python
#Análisis estadístico
flights.describe()
```

```python
# Pivotear columnas de dataset
flights = flights.pivot("month","year","passengers")
```

## Heatmap

```python
sns.heatmap(flights)
```

### Agregar etiquetas de datos

```python
sns.heatmap(flights, annot=True)
```

## Ajuste de etiqueta de datos

```python
sns.heatmap(flights, annot=True, fmt="d")
```

## Ajustar espacio entre elementos de gráfico

```python
sns.heatmap(flights, annot=True, fmt="d", linewidths=1.5)
```

## Ajustar valores mínimo y máximo de gráfico

```python
sns.heatmap(flights, annot=True, fmt="d", linewidths=1.5, vmin=150, vmax=650)
```

## Cambiar paleta de colores

```python
sns.heatmap(flights, annot=True, fmt="d", linewidths=1.5, vmin=150, vmax=650, cmap="RdBu")
```

```python
sns.heatmap(flights, annot=True, fmt="d", linewidths=1.5, vmin=150, vmax=650, cmap="summer")
```

```python
sns.heatmap(flights, annot=True, fmt="d", linewidths=1.5, vmin=150, vmax=650, cmap="winter_r")
```

## Asignar centro de color del centro del grafico

#### Ubicar celda central para indicar en la instrucción.

```python
sns.heatmap(flights, center=flights.loc["June", 1954], annot= True, fmt="d")
```

### Retirar escala de gráfico

```python
sns.heatmap(flights, center=flights.loc["June", 1954], annot= True, fmt="d", cbar=False)
```
