```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
#Generar lista de números aleatorios
num = np.random.randn(150)
```

## Graficar Histograma - distribución

```python
sns.set(style="darkgrid")
sns.distplot(num)
```

### Cambio de color

```python
sns.set(style="darkgrid")
sns.distplot(num, color ="red")
```

### Crear etiqueta a gráfica

```python
label_dist = pd.Series(num, name="variable_x")
```

```python
sns.distplot(label_dist)
```

### Cambiar Histograma a horizontal

```python
sns.distplot(label_dist, vertical = True , color = "red")
```

### Eliminar histograma de gráfica

```python
sns.distplot(label_dist, vertical = True , color = "red", hist=False)
```

### Agregar RUG - Datapoint

```python
sns.distplot(label_dist , color = "red", hist=False, rug = True)
```
